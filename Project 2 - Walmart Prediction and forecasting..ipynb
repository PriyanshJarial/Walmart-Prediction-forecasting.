```python
# Checking Outliers
col = list(df.columns)
for co in col:
    if df[co].dtypes == 'int64' or df[co].dtypes == 'float64':
        plt.boxplot(df[co])
        plt.xlabel(co)
        plt.show();
```

```python
# Removing the Outliers for Temperature.
Q1 = df['Temperature'].quantile(0.25)
Q3 = df['Temperature'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['Temperature']>= Q1-1.5*IQR)&(df['Temperature']<=Q3+1.5*IQR)]
```

```python
# Removing the Outliers for Unemployment.
Q1 = df['Unemployment'].quantile(0.25)
Q3 = df['Unemployment'].quantile(0.75)
IQR = Q3-Q1
df=df[(df['Unemployment']>=Q1-1.5*IQR)&(df['Unemployment']<=Q3+1.5*IQR)]
```

```python
# Again checking for the Outliers.
col=list(df.columns)
for co in col:
    if df[co].dtypes=='int64' or df[co].dtypes=='float64':
        plt.boxplot(df[co])
        plt.xlabel(co)
        plt.show();
```

```python
# Getting unique columns for each Store and Holiday_Flag. 
df1= pd.get_dummies(df,columns=['Store','Holiday_Flag'])
df1
```

```python
# importing liberary for Train and Test with models.
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
x=df1.drop(['Date','Weekly_Sales','Week','Month','Year'],axis=1)
y=df1['Weekly_Sales']
```

```python
# Spliting data into Training on 80% and Testing on 20%.
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=1)
```

```python
# Using  the LinearRegression Model and fitting.
lin = LinearRegression()
modf1 = lin.fit(x_train,y_train)
```

```python
y_pr1 = modf1.predict(x_test)
```

```python
print( 'LinearRegression()\n')

print('Mean Square Error: ',mean_squared_error(y_test,y_pr1))
print('Mean Absolute Error: ',mean_absolute_error(y_test,y_pr1))
print('Root Mean Square Error: ',np.sqrt(mean_squared_error(y_test,y_pr1)))
print('R2 Score: ',r2_score(y_test,y_pr1))
```

```python
# Using the RandomForestRegressor Model and fitting. 
ran = RandomForestRegressor(n_estimators=100)
modf2 = ran.fit(x_train,y_train)
```

```python
y_pr2 = modf2.predict(x_test)
```

```python
print('RandomForestRegressor()\n')
print('Mean Square Error: ',mean_squared_error(y_test,y_pr2))
print('Mean Absolute Error: ',mean_absolute_error(y_test,y_pr2))
print('Root Mean Square Error: ',np.sqrt(mean_squared_error(y_test,y_pr2)))
print('R2 Score: ',r2_score(y_test,y_pr2))
```

```python
df.set_index('Date',inplace=True)
```

```python
s20 = df[df['Store']==20].sort_values('Date')
s20 =s20.groupby('Date')['Weekly_Sales'].sum()
s20 = pd.DataFrame(s20)
s20
```

```python
plt.plot(s20['Weekly_Sales'])
```

```python
from statsmodels.tsa.stattools import adfuller
p = adfuller(s20)
if p[1] > 0.05:
    print('Non Stationary')
else:
    print('Stationary')
print(p)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
seasonal_decompose(s20,period=36).plot();
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
plot_pacf(s20)
plot_acf(s20)
```

```python
from pmdarima import auto_arima
auto_arima(s20,trace=True)
```

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(s20,order=(0,0,1))
mf = model.fit()
s20['AR_predict']=mf.predict(start=pd.to_datetime('2011-11-11'))
```

```python
s20[['Weekly_Sales','AR_predict']].plot()
```
